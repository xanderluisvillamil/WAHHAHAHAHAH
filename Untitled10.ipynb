```python
subjects = ["Math", "English", "CS", "Adtech", "PEHM", "VALED", "Filipino", "SocSci", "IS"]
total_grades = 0
count = len(subjects)


print("Grade Calculator Initialized")


for subject in subjects:
    while True:
        try:
            grade = float(input(f"Enter grade for {subject}: "))
            if 0 <= grade <= 100:
                break
            else:
                print("Invalid input. Please enter a grade between 0 and 100:")
        except ValueError:
            print("Invalid input. Please enter a numeric grade:")


    total_grades += grade


average = total_grades / count


print("Calculation Complete")
print(f"Final average is: {average:.2f}")


if average >= 75:
    print("Status: PASSED")
else:
    print("Status: FAILED")
```
